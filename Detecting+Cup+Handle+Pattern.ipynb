```python
import yfinance as yf
import mplfinance as mpf
import matplotlib.pyplot as plt
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
tsla=yf.Ticker("TSLA").history('5y')
```

```python
mpf.plot(tsla, savefig='img.png')
```

```python
mpf.plot(tsla['2021-03-01':'2021-04-30'])
```

```python
#Download data 
tsla.to_csv('tsla.csv')
data=pd.read_csv('tsla.csv')
data.info()
```

```python

data.Date=pd.to_datetime(data.Date)  #change 
data=data.set_index('Date')
```

```python
mpf.plot(data)
```

```python
mpf.plot(data['2021'])
```

```python
mpf.plot(data['2020'], type='line')
```

```python
mpf.plot(data['2021'], type='candle')
```
